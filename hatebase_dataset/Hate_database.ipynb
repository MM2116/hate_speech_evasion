```python
#ORIGINAL DATASET EXTRCTION
import json
import pprint

hate_dict = {}
for i in range(1,16):
    #Sequentital file handling
    file_name = "page_"+ str(i)
    hate_dict[file_name] = []
    print(file_name)
    with open(file_name) as json_file:
        #loading each json file
        data = json.load(json_file)
        for result in data['result']:
            word = result['term']
            hateful_meaning= result['hateful_meaning']
            offensive_score=result['average_offensiveness']
            variant_of=result['variant_of']
            occurances=result['number_of_sightings']
            word_json = {'word': word, 'hateful_meaning': hateful_meaning, \
                         'offensive_score': offensive_score, 'variant_of': variant_of, 'occurances': occurances}
            hate_dict[file_name].append(word_json)
#print(hate_dict)
print(json.dumps(hate_dict, indent = 4))
```

```python
#WRITING results from particular file
with open("hate_json","w") as hate_json_file:
    hate_json_file.write(json.dumps(hate_dict, indent = 4))
```
